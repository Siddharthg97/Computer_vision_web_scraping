```python
# import pandas lib as pd
import pandas as pd
 
# read by default 1st sheet of an excel file
dataframe1 = pd.read_excel('Book3.xlsx')
 
print(dataframe1)
```

```python
from bs4 import BeautifulSoup as bs 
import requests
import json
import pandas as pd
import io
```

```python
passlist=[]
contentlist=[]
passnflist=[]
for i in range(500):
    print(i)
#     i=4
    cntr=dataframe1.iloc[[i],[2]]
    cntr1=cntr.to_string(index=False).split('\n')
    cntrf=cntr1[1].strip()
    print(cntrf)
    coden=dataframe1.iloc[[i],[1]]
    coden1=coden.to_string(index=False).split('\n')
    codenf=coden1[1].strip()
    codenf=codenf.replace("[",'')
    codenf=codenf.replace("]",'')
    codenf=codenf.split(', ')
    #print(codenf)

    passn=dataframe1.iloc[[i],[0]]
    passn1=passn.to_string(index=False).split('\n')
    #print(codenf)
    passnf=passn1[1].strip()
    passnf=passnf.replace("[",'')
    passnf=passnf.replace("]",'')

    passnf=passnf.split(',')
    #print(len(passnf))
    #passnf=passnf.replace()
    #passnf=passnf.strip()
    #print(len(passnf))
    #print(passnf)
    if len(passnf)==2:
        print("ok")
        try:
            codenx=codenf.index("'PASS'")
        except:
            print("PASS not found")
            passnflist.append([i])
            continue
    #     print(codenx)
        passnf1=passnf[codenx]
    #     print(passnf[codenx])
    else:
        passn=dataframe1.iloc[[i],[0]]
        passn1=passn.to_string(index=False).split('\n')
        #print(codenf)
        passnf=passn1[1].strip()
        passnf=passnf.replace("[",'')
        passnf=passnf.replace("]",'')
        passnf1=passnf
    passnf1=passnf1.replace("'",'')
    passnf1=passnf1.strip()
    print(passnf1)

    URL = 'https://eservices.imi.gov.my/myimms/FomemaStatus' 

    #print(passnf1)
    data = {
        'txnDetail': '',
        'SEARCH_IND': '',
        'MVP_NAT_CD': cntrf,
        'CurrLang': 'ms',
        'MVP_DOC_NO': passnf1,
        'MVP_NAT_CD1': 'abc',
        'MVP_NAT': cntrf,
        'MVP_NM': '',
        'MVP_NAT_CD2': '',
        'MVP_NAT1': '',
        'searchStatusFomema': 'Carian',
#         '_sourcePage': '1Tu_-xNmZhajTsv9kmp8gDF7a_BOL6gZtsdKLMIAsLNdSbt1vmia7BnDveO2yIACF_hVVHeDnoo=',
#         '__fp': 'k1N1_PaV8FU=',
    }


    s = requests.session()
    #s.trust_env=False

    #response = http.get(URL, data=data)
    response = s.post(URL, data=data)
    #print(response.content)
    soup = bs(response.content, "html.parser")
    #response.encoding = 'utf-8'
    #j=response.json
    #jsonda=response.json
    text = soup.get_text(' | ',strip=True)
    #print(text)
    delim=" | "
    ntext=text.split(delim)
    try:
        inx=ntext.index('1')
    except:
        print("Content is not found")
        contentlist.append([i])
        continue
    #print(inx)
    output=ntext[inx:inx+5]
    # print(ntext)
    #print(output)
    output1=output+[i]+[cntrf]
#     print(i)
#     print(output1)
    passlist.append(output1)
    #passlist.append([i])
passdf=pd.DataFrame(passlist)
passdf.columns=['Idx','ID NO','Name','DOB','Gender','Data_idx','Country']
print(passdf)
print(contentlist)
print(passnflist)
```

```python
passdf.to_csv("customsoutput.csv")
```
